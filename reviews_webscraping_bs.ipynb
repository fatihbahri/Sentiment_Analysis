## Libraries and Definations

```python
#import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

This defination is cleaning the emojis from reviews

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U0001F1F2-\U0001F1F4"  # Macau flag
        u"\U0001F1E6-\U0001F1FF"  # flags
        u"\U0001F600-\U0001F64F"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U0001F1F2"
        u"\U0001F1F4"
        u"\U0001F620"
        u"\U0001F9D0"
        u"\U0001F917"
        u"\U0001F970"
        u"\u200d"
        u"\u2640-\u2642"
        "]+", flags=re.UNICODE)
```

```python
def replace_text(rew):
    rew = rew.replace(",", "") #remove comma signs (comma signs cause a commotion in csv files)
    rew = rew.replace(";", "")
    rew = rew.replace(".", "")
    rew = rew.replace("\n", " ")
    rew = rew.replace("\t", " ") #remove other blanks
        #or use re.sub() method #
        #rew = re.sub('[^a-zA-Z]', ' ','rew')
    rew = rew.lower() #all letters will be lowercase
    return rew
```

## Scraping Reviews and Scores from page

Scraping scores with for and try-except loops.

Try-Except is used to catch the div's noninclusive scores

"lim" variable is used to define all reviews and "star2" is used for all scores in the page

```python
lim = soup.find_all("p", class_='review-text mb-2 mt-5')
#star = soup.find_all("div", class_='mb-2')
star2 = soup.find("ul", {"class":"timeline-property mt-5"}).find_all("li")
stars = []
for rates in star2:
    try:
        yildiz = rates.find("div", {"class":"timeline-data"}).find("div", {"class":"media-body"}).find("div", {"class":"mb-2"}).find("i")
        star = 1 if int(yildiz['class'][2][-1]) > 3 else 0
        stars.append(star)
    except:
        continue
```

Labeling scraped reviews as:
_________
0: Negative Reviews

1: Possitive Reviews

```python
stars
```

```python
reviews = []
```

```python
tr_to_en = str.maketrans("çğıöşüÇĞİÖŞÜ", "cgiosuCGIOSU")
```

```python
for rews in lim:
    rew = rews.text #get text
    rew = rew.strip() #remove blanks
    rew = rew.translate(tr_to_en) #replace turkish characters to latin.
    rew = emoji_pattern.sub(r'', rew) #remove emojis
    rew = replace_text(rew)
    reviews.append(rew) #save to reviews list
```
